6. Use the Boolean mask from 5 to filter `vehicles` down to only manual transmission vehicles. Assign the result to the variable `manual`.

7. We've created a count of manual transmission vehicles by year. Add code to sort the result by year from earliest to latest. What do you notice?

```python
vehicles = manual_vehicles.sort_values (by = 'year', ascending=True)
vehicles.head()
```

8. Create a Boolean mask that is `True` when `vehicle_type` is `suv` or `pickup`. Filter `vehicles` down to only the `True` rows.

```python
is_suv = vehicles ['vehicle_type'] == 'suv' 
is_pickup = vehicles ['vehicle_type'] == 'pickup'
is_suv_or_pickup = vehicles [is_suv | is_pickup]
is_suv_or_pickup.head()
```

9. Create a Boolean mask that is `True` when `vehicle_type` is neither `suv` nor `pickup`. Filter `vehicles` down to only the `True` rows.

```python
is_suv = vehicles ['vehicle_type'] == 'suv' 
is_pickup = vehicles ['vehicle_type'] == 'pickup'
is_suv_and_pickup = vehicles [is_suv & is_pickup]
not_suv_nor_pickup = ~is_suv_and_pickup
non_suv_or_pickup_vehicles = vehicles [not_suv_nor_pickup]
not_suv_nor_pickup.head()
```

10. Create a Boolean mask that is `True` whenever `fuel` is `hybrid electric` and `best_mpge` is below average (less than 40). Filter `vehicles` down to only the rows that are `True`.

```python
is_hybrid = vehicles ['fuel'] == 'hybrid electric'
poor_mpge = vehicles ['best_mpge'] <= 40
is_hybrid_with_low_mpge = vehicles [is_hybrid & poor_mpge]
is_hybrid_with_low_mpge.head(25)
```
